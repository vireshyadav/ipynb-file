```python
from autoscraper import AutoScraper
```

```python
gitar = ("https://www.amazon.in/s?k=gitar&crid=3G2MFE29KVT11&sprefix=gitar%2Caps%2C303&ref=nb_sb_noss_2")

wanted_list = ("7,077","Vault EA40 41 inch Premium Solid Spruce-Top Cutaway Acoustic Guitar","Product details"," 377")
```

```python
scraper = AutoScraper()
result = scraper.build(gitar,wanted_list)
print(result)
```

```python
scraper.get_result_similar(gitar,grouped = True)
```
